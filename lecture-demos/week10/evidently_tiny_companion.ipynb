```python
import numpy as np
import pandas as pd

np.random.seed(42)

# Simple synthetic data
df_train = pd.DataFrame({
    'age': np.random.normal(30, 5, 300),
    'income': np.random.normal(8, 2, 300),
    'city': np.random.choice(['A', 'B', 'C'], size=300, p=[0.4, 0.35, 0.25])
})

df_prod = pd.DataFrame({
    'age': np.random.normal(38, 7, 300),
    'income': np.random.normal(6.5, 2.5, 300),
    'city': np.random.choice(['A', 'B', 'C'], size=300, p=[0.2, 0.2, 0.6])
})

df_train.head()
```

## Why use Evidently?

Instead of manually running different tests feature by feature, Evidently can:

- detect drift on many features at once
- choose suitable checks for numeric and categorical columns
- generate an HTML report

```python
from evidently.report import Report
from evidently.metric_preset import DataDriftPreset

report = Report(metrics=[DataDriftPreset()])
report.run(reference_data=df_train, current_data=df_prod)
report.save_html('drift_report.html')
print('Saved drift_report.html')
```
